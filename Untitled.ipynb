```python
from sklearn import datasets 
import pandas as pd 
import numpy as np
from math import sqrt
import matplotlib.pyplot as plt

iris = datasets.load_iris()
iris = pd.DataFrame(data= np.c_[iris['data'], iris['target'] + 1],
                     columns= iris['feature_names'] + ['class'])
iris
```

# (a) Centroids

```python
iris_sentosa = iris[iris['class'] == 1]
iris_versi = iris[iris['class'] == 2]
iris_verginica = iris[iris['class'] == 3]

print("SENTOSA")
print(iris_sentosa.mean(axis = 0))
print("VERSICOLOR")
print(iris_versi.mean(axis = 0))
print("VIRGINICA")
print(iris_verginica.mean(axis = 0))
```

# (b) Farthest Point 

```python
def euclid(x, y):
    return sqrt(sum((i - j)**2 for i, j in zip(x, y)))
# Sentosa
sentosa = iris_sentosa.values.tolist()
sentosa_mean = [5.006, 3.428, 1.462, 0.246, 1.000]
# x = zip(sentosa[1], sentosa_mean)
# print(list(x))
max_dist = -9999
vec = []
for i in sentosa:
    distance = euclid(i, sentosa_mean)
    if distance > max_dist:
        max_distance = distance
        vec = i
print(max_distance, vec)  

#VersiColor
versi = iris_versi.values.tolist()
versi_mean = [5.936, 2.770, 4.260, 1.326, 2.000]
max_dist = -999
for i in versi:
    distance = euclid(i, versi_mean)
    if distance > max_dist:
        max_distance = distance
        vec = i
print(max_distance, vec)  

#Virginica
verginica = iris_verginica.values.tolist()
verginica_mean = [6.558, 2.974, 5.552, 2.026, 3.000]
max_dist = -999
for i in verginica:
    distance = euclid(i, verginica_mean)
    if distance > max_dist:
        max_distance = distance
        vec = i
print(max_distance, vec)  
```

# (a) Generalization
#### In pattern recognition it is the idea to use a number of properties(features) of a particular object or a class in such a way that allows to separate the objects or the classes from the others. For example like in the Iris dataset we have petals and sepals dimensions of three different plants which can be used to find a pattern or a threshold where we can distinguish between the three classes based on this information.
# (b) Intra Class Variations 
#### It is generally referred to as the variations of feature samples within a particular class. For example lets take grapes and apples, apples are of different sizes(ranging from small to too large), shapes and colors, but grapes have very less variations compared to apples as grapes are generally small and are colored as shades of green and yellow.
# (c) Decision Boundary
#### It is basically a rule that can be used to separate out different classes based on the features of different class. For example if we are trying to separate out sea bass and salmon based of there sizes or colors or number of fins, we can use our features space to determine a decision boundary that can be used to separate out sea bass and salmon.
# (d) Reinforcement Learning
#### It is an area in machine learning where agents must take appropriate actions in an environment in order to maximize their reward. For example we take a game of chess where the agent has to play the right steps in order to win the game against the player.
# (e) Invariant Representation
#### In our context itâ€™s the representation of features so that they are invariant (not affected) by any other factors that are present while we are recording them. For example if we are separating out sea bass and salmon in a conveyor belt, so the position of the fishes on the conveyor belt should be in variant to the orientation of the fishes on the belt.

```python
fig, axs = plt.subplots(1,1)

data[data["class"] == 1].plot.scatter(x="x1", y="x2", c="red", marker="o", ax=axs, label="Class 1")
data[data["class"] == 2].plot.scatter(x="x1", y="x2", c="blue", marker="x", ax=axs, label="Class 2")

x1 = np.linspace(-4, 15, 1000)
x2 = [15 - i for i in x1]
plt.plot(x1, x2)

pos_w1 = 0
neg_w1 = 0
pos_w2 = 0
neg_w2 = 0
for i, j in data.iterrows():
    if j['x1'] + j['x2'] - 15 < 0:
        if j["class"] != 1:
            neg_w1 += 1

    elif j['x1'] + j['x2'] - 15 > 0: 
        if j["class"] != 2:
            neg_w2 += 1

print("Error rate: {}%".format(((neg_w1 + neg_w2)/200)*100))
```
